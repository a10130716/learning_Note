## 3.成果呈現

```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/will/Downloads/20歲以上國人糖尿病盛行率.csv', encoding='big5')
```

```python
df
```

```python
df2 = pd.read_csv('/Users/will/Downloads/20歲以上國人高血壓盛行率.csv', encoding='big5')
```

```python
df2
```

```python
df.columns = [df.columns[0]]+list(df.columns[1:]+'(高血壓)')
```

```python
df
```

```python
df2.columns = [df2.columns[0]]+list(df2.columns[1:]+'(糖尿病)')
```

```python
df2
```

```python
test = pd.merge(df,df2, on='年份別')
```

```python
pd.melt(test,['年份別'], var_name='疾病分類', value_name='數值')
```
